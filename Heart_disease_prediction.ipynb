importing the dependencies ( libraries and functions )


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data collection and processing

```python
1 #loading the CSV data to a panda DataFrame
heart_data = pd.read_csv('/content/heartData.csv')
```

```python
#first 5 rows of the data
print(heart_data.head())
```

```python
#printing last 5 cols of the data
heart_data.tail()
```

```python
#number of rows and cols in data set
heart_data.shape
```

```python
#information about data
heart_data.info()
```

```python
# checking for null vallues in data
heart_data.isnull().sum()
```

```python
#statistical measure about the data
heart_data.describe()
```

```python
# cheking the distribution of target variable
heart_data['target'].value_counts()
```

1 -> diseased heart

0 -> healthy heart

Splitting Target and Features

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting data into Training data and Testing data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.3,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic regression

```python
model = LogisticRegression()
```

```python
#Training the logisticRegression model with training data
model.fit(X_train,Y_train)
```

 **Model evaluation**

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy )
```

```python
# accuracy on Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data : ',test_data_accuracy )
```

Building a Predictive System

```python
input_data = (59,1,3,130,318,0,0,120,1,1,2)

#change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
  print('The person does not have a heart disease')
else:
  print('The person has heart disease')
```
